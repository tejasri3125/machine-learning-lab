```python
import numpy as np 
import pandas as pd 
import scipy as sp 
import matplotlib as mpl 
import seaborn as sns 

df = pd.read_csv("Heart_Disease_Prediction.csv")
df.head()
```

```python
df.columns = df.columns.str.strip().str.lower()
df.columns
```

```python
for col in df.columns:
    print(repr(col))
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ', '_')
)
print(df.columns)
```

```python
mean_age = df['age'].mean()
median_age = df['age'].median()
mode_age = df['age'].mode().iloc[0]
variance_age = df['age'].var()
std_dev_age = df['age'].std()

print("Mean Age:", mean_age)
print("Median Age:", median_age)
print("Mode Age:", mode_age)
print("Variance Age:", variance_age)
print("Standard Deviation Age:", std_dev_age)
```

```python
cov_age_maxhr = df[['age', 'max_hr']].cov().iloc[0, 1]
print("Covariance between Age and Max HR:", cov_age_maxhr)
```

```python
df.groupby('heart_disease')['age'].mean()
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
print(df.isnull().sum())
```

```python
df = df.fillna(df.mean(numeric_only=True))
```

```python
print("\n--- Missing Values Processed ---")
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
```

```python
if len(numeric_cols) < 2:
    print("ERROR: Dataset must contain at least one feature and one target.")
    exit()

```

```python
target_col = numeric_cols[-1]
```

```python
feature_cols = numeric_cols[:-1]
```

```python
X = df[feature_cols]
```

```python
y = df[target_col]
```

```python
print(f"\nFeatures (X): {feature_cols}")
```

```python
print(f"Target (y): {target_col}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=27
)

```

```python
print("\n--- Data Split Complete (70% Train, 30% Test) ---")

```

```python
model = LinearRegression()
model.fit(X_train, y_train)

print("\n--- Model Training Complete ---")

```

```python
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

```

```python
print("\n--- Model Evaluation ---")
```

```python
print(f"Mean Absolute Error: {mae:.2f}")
```

```python
print(f"R2 Score: {r2:.2f}")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title(f"Actual vs Predicted ({target_col})")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(x=y_test, y=y_pred, scatter_kws={'alpha': 0.5})
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Regression Results: Actual vs Predicted")
plt.show()

print("\n--- Successfully Completed Linear Regression ---")
```
